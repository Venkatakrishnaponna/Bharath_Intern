```python
from sklearn.metrics import r2_score
```

```python
y_actual=df.area
y_actual
```

```python
y_predict=reg.predict(df[['price']])
y_predict
```

```python
rSquare=r2_score(y_actual,y_predict)
rSquare
```

```python
Accuracy=rSquare*100
Accuracy
```

```python
Loss=100-Accuracy
Loss
```
